# **1. User Entity**




Stores account-level information for each Netflix subscriber.

Attribute	Description

---



UserID (PK)	Unique identifier for each user account

---


FullName	Full name of the account holder

---


Email	User email address for login and communication

---


PasswordHash	Encrypted password for account security

---


PhoneNumber	Contact phone number

---


Country	User's country of residence

---


SubscriptionID (FK)	Linked subscription plan for the account

---


RegistrationDate	Date when the account was created

---




# **2. Profile Entity**


Each user account can have up to 5 separate profiles, each with independent preferences.

---


Attribute	Description

---


ProfileID (PK)	Unique identifier for each profile

---


UserID (FK)	References the parent user account

---


ProfileName	Display name for the profile (e.g., Mom, Kids)

---


AgeCategory	Kids or Adult profile type

---


LanguagePreference	Preferred content and subtitle language

---


AvatarURL	URL to the profile picture/avatar

# **3. Subscription Entity**

Stores Netflix subscription plan details and defines access privileges.

---


Attribute	Description

---


SubscriptionID (PK)	Unique identifier for each subscription plan

# **4. Content Entity**

Stores all movie and TV show metadata available on the platform.

---


Attribute	Description

---

Attribute	Description

# **SQL Schema Implementation**

CREATE TABLE Users (

    UserID        INT PRIMARY KEY,

    FullName      VARCHAR(100),

    Email         VARCHAR(100) UNIQUE NOTNULL,

    PasswordHash  VARCHAR(255) NOT NULL,

    PhoneNumber   VARCHAR(20),

    Country       VARCHAR(50),

    SubscriptionID INT,
    
    RegistrationDate DATE
);


CREATE TABLE Subscription (
  
    SubscriptionID INT PRIMARY KEY,

    PlanName       VARCHAR(50),

    MonthlyPrice   DECIMAL(10,2),

    MaxDevices     INT,

    VideoQuality   VARCHAR(20)
);


CREATE TABLE Profiles (
  
    ProfileID          INT PRIMARY KEY,

    UserID             INT,

    ProfileName        VARCHAR(100),

    AgeCategory        VARCHAR(20),

    LanguagePreference VARCHAR(50),

    AvatarURL          VARCHAR(255),

    FOREIGN KEY (UserID) REFERENCES Users(UserID)
);



CREATE TABLE Content (

    ContentID   INT PRIMARY KEY,

    Title       VARCHAR(255),

    Genre       VARCHAR(100),

    ReleaseYear INT,

    Duration    INT,

    Language    VARCHAR(50),

    ContentType VARCHAR(20),

    Description TEXT,
    
    IMDbRating  DECIMAL(3,1)
);


CREATE TABLE WatchHistory (

    WatchID          INT PRIMARY KEY,

    ProfileID        INT,

    ContentID        INT,

    WatchDate        DATE,

    WatchDuration    INT,

    CompletionStatus VARCHAR(20),

    FOREIGN KEY (ProfileID) REFERENCES Profiles(ProfileID),
    
    FOREIGN KEY (ContentID) REFERENCES Content(ContentID)
);



CREATE TABLE Rating (

    RatingID    INT PRIMARY KEY,

    ProfileID   INT,

    ContentID   INT,

    RatingValue DECIMAL(3,1),

    RatingDate  DATE,

    FOREIGN KEY (ProfileID) REFERENCES Profiles(ProfileID),
    
    FOREIGN KEY (ContentID) REFERENCES Content(ContentID)
);


CREATE TABLE Watchlist (

    WatchlistID INT PRIMARY KEY,

    ProfileID   INT,

    ContentID   INT,

    AddedDate   DATE,

    FOREIGN KEY (ProfileID) REFERENCES Profiles(ProfileID),
    
    FOREIGN KEY (ContentID) REFERENCES Content(ContentID)
);


CREATE TABLE Device (

    DeviceID   INT PRIMARY KEY,

    UserID     INT,

    DeviceType VARCHAR(50),

    DeviceName VARCHAR(100),

    LastLogin  TIMESTAMP,
    
    FOREIGN KEY (UserID) REFERENCES Users(UserID)
);
