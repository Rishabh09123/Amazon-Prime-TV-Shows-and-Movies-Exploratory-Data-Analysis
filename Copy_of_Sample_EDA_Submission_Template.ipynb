## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd

# Load datasets
titles = pd.read_csv('/content/drive/MyDrive/titles.csv')
credits = pd.read_csv('/content/drive/MyDrive/credits.csv')

# Check basic info
titles.info()
credits.info()

# Handle missing values
# Convert imdb_score to numeric, coercing errors to NaN
titles['imdb_score'] = pd.to_numeric(titles['imdb_score'], errors='coerce')
titles = titles.dropna(subset=['title', 'imdb_score']) # Drop NaNs after conversion
titles['age_certification'] = titles['age_certification'].fillna('Unknown')

# Convert data types if needed
titles['release_year'] = titles['release_year'].astype(int)

# Split genres for analysis
titles['genres'] = titles['genres'].str.replace("[\\[\\]']", "", regex=True)
titles['genres'] = titles['genres'].str.split(',')

# Explode genres
titles_exploded = titles.explode('genres')

# Merge datasets
df = pd.merge(titles, credits, on='id', how='left')

# Final dataset check
df.head()
````

### What all manipulations have you done and insights you found?

Cleaned the dataset by removing missing and irrelevant values
Filled missing age certifications for better consistency
Converted columns into proper formats for analysis
Split and expanded genres to analyze each genre separately
Merged titles and credits dataset for deeper insights

**Insights:**

Data had missing values mainly in ratings and certifications
Some columns (like genres) needed transformation to be useful
After cleaning, dataset became structured and ready for analysis

````python
# Chart - 1 visualization code
# Chart - 1: Top 10 genre  (bar chart)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/titles.csv')

# Normalize case before exploding
titles_exploded = (
    df.assign(genres=df['genres'].str.lower().str.split(', '))
      .explode('genres')
)

titles_exploded['genres'] = titles_exploded['genres'].str.strip()

genre_count = titles_exploded['genres'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=genre_count.values, y=genre_count.index, palette='viridis')
plt.title("Top 10 Genres on Amazon Prime")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
````

#### Chart - 3

````python
# Chart - 3 visualization code
# Chart - 3  │  HEATMAP  │  Average IMDb Score by Genre × Content Type

titles_exploded = titles.explode('genres')
titles_exploded['genres'] = titles_exploded['genres'].str.lower().str.strip()

pivot = (
    titles_exploded
    .groupby(['genres', 'type'])['imdb_score']
    .mean()
    .unstack('type')
    .dropna(how='all')
)

# Keep only the top genres by overall count to avoid clutter
top_genres = titles_exploded['genres'].value_counts().head(12).index
pivot = pivot.loc[pivot.index.isin(top_genres)]

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    pivot,
    annot=True, fmt=".1f",
    cmap='YlOrRd',
    linewidths=0.5,
    ax=ax,
    cbar_kws={'label': 'Avg IMDb Score'}
)
ax.set_title("Chart 3 – Avg IMDb Score: Genre × Content Type", fontsize=13, pad=12)
ax.set_xlabel("Content Type")
ax.set_ylabel("Genre")
plt.tight_layout()
plt.savefig('chart3_heatmap_genre_type_imdb.png', dpi=150)
plt.show()
````

##### 1. Why did you pick the specific chart?

A heatmap is ideal for visualising a matrix of two categorical variables

against a numeric measure. It lets viewers instantly spot which genretype

combinations have the highest/lowest audience ratings—something a bar chart

or scatter plot would struggle to convey compactly.


##### 2. What is/are the insight(s) found from the chart?

• Documentation and history genres tend to score highest, especially for MOVIES.

 • Horror consistently scores below 6 for both MOVIEs and SHOWS.

 • SHOWs generally score ≥ 0.3 pts higher than MOVIEs in the same genre.

#### Chart - 4

````python
# Chart - 4 visualization code
# Chart - 4  │  STACKED BAR  │  Top 8 Production Countries – MOVIE vs SHOW

top_countries = df['production_countries'].value_counts().head(8).index
df_c = df[df['production_countries'].isin(top_countries)]
country_type = df_c.groupby(['production_countries', 'type']).size().unstack(fill_value=0)

country_type.plot(kind='bar', stacked=True, figsize=(9, 5), color=['#1f77b4', '#ff7f0e'])
plt.title("Top 8 Production Countries: Movies vs Shows")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A stacked bar chart efficiently shows both the total content volume per country AND how that volume splits between Movies and Shows—two dimensions in a single, easy-to-read graphic.

##### 2. What is/are the insight(s) found from the chart?

• The US dominates (~5 000 titles)—roughly 5× the next largest producer (IN).

• India (IN) and GB are strong contributors, but their mixes lean more toward Movies than the US does.

• Canada (CA) and France (FR) are mid-tier contributors, mostly Movies.

## **5. Solution to Business Objective**

#### What do you suggest the client to achieve Business Objective ?
Explain Briefly.

Based on the analysis of Amazon Prime Video's content catalog, the following solutions are recommended:

1. Focus on High-Performing Genres
Documentary and history content consistently scores the highest ratings. Amazon Prime should acquire or produce more content in these genres as they directly improve viewer satisfaction.

2. Produce More TV Shows
TV Shows score higher than Movies in most genres. More original show productions means higher watch time and better subscriber retention compared to one-time movie views.

3. Grow Regional Content
India is the second largest content contributor. Increasing investment in Indian original content can drive domestic growth and reduce over-reliance on US licensed content.

4. Remove Low-Performing Titles
A large number of low-rated titles are hurting overall catalog quality. Removing consistently underperforming content will improve recommendation accuracy and user trust.

5. Add Classic Content
Very few titles exist from before 1990. Adding selected classic films can attract an older audience segment that current content does not serve.
